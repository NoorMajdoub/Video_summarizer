```python
import cv2
import os
def extract_frame_at_timestamp(video_path, timestamp_seconds, output_path):
    video = cv2.VideoCapture(video_path)
    fps = video.get(cv2.CAP_PROP_FPS)

    frame_number = int(timestamp_seconds * fps)
    video.set(cv2.CAP_PROP_POS_FRAMES, frame_number)

    success, frame = video.read()
    if success:
        cv2.imwrite(output_path, frame)
    else:
        print(f"Failed to get frame at {timestamp_seconds} seconds. why we dont know")

    video.release()

```

```python
import cv2

# Load image
image = cv2.imread('/kaggle/working/cropped_code_section3.png')

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply adaptive threshold to isolate text
thresh = cv2.adaptiveThreshold(
    gray, 255, 
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
    cv2.THRESH_BINARY, 11, 2
)

# Optional: Invert if text is light on dark
inverted = cv2.bitwise_not(thresh)

# Save and run OCR
cv2.imwrite('preprocessed_code3.png', inverted)
```
